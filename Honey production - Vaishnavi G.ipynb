# 4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.


```python
sns.displot(x=df['priceperlb'],kind='hist',bins=30,kde=True)    #using displot, plotting the distribution of variable
plt.show()                                                            
```

# 7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')


```python
plt.figure(figsize=(12,5))   #using pairplot, visualizing the relationship of multiple pairs of variables
sns.pairplot(df[['numcol', 'yieldpercol','totalprod', 'prodvalue','year']],hue='year')
```

```python
df[['prodvalue']].kurtosis()
```

```python
## FINDINGS:

#   'numcol','totalprod','prodvalue' are positively skewed.
#   'yieldpercol' is normally distributed graph
#   So the 'numcol','totalprod','prodvalue' are positively skewed with each other.
#   'product value' has tall and sharp peak and is leptokurtic
#   product value and numcol,totalproduct and numcol is related Linearly 
```
